## What this notebook does

Country-block bootstrap (B=200 iterations) for **both regression and ML models**:

| Phase | What | Duration |
|-------|------|----------|

| 2 (NB6) | Fit Models 3a / 3b on each bootstrap dataset, compute 95% CIs | ~2 min |
| 3 (NB5) | Fit LASSO / Ridge / EN / RF on each bootstrap dataset, compute 95% CIs | ~40 min |
